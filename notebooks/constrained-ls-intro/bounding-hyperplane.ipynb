```python
DIR = '../../content/images/constrained-ls-intro/'
```

```python
n_planes = 5

x = linspace(-1, 1, 100)

plot(x, x**2)
for q in linspace(-.75, .75, n_planes):
    plot(x, (x-q)*(2*q) + q**2, color="gray")
    scatter(q, q**2, facecolors="none", edgecolors="r")

ylim([-.25, 1])
xlim([-1, 1])
axis('off')
savefig(DIR+'bounding-hyperplanes.png', dpi=300, bbox_inches='tight')
```
