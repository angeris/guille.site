```python
def phi(x):
    return 1/(1+np.exp(x))
```

```python
x = linspace(-3, 3, 300)
C_space = logspace(-1, 2, 5)

for C in C_space:
    plot(x, phi(C*x), label='$C = {:.2f}$'.format(C))
    
title(r'$\phi(\cdot)$ as a function of curvature parameters')
xlabel('$x$')
ylabel('$\phi(x)$')
ylim([-.01, 1.01])
legend()
sns.despine()
savefig('../../content/images/path-optimization-1/phi_curvature.png')
```
